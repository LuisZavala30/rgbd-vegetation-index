```python
import pandas as pd
df_final = pd.DataFrame()

#df_final['Region'] = None
df_final['R'] = None
df_final['G'] = None
df_final['B'] = None
df_final['Depth'] = None
df_final['NDVI'] = None

#Region_list = []
R_list = []
G_list = []
B_list = []
Depth_list = []
NDVI_list = []

```

```python
import os
#from PIL import Image
import cv2 
import numpy as np


## Specify the paths that contains the files with the cropped data
input_R_path =""
input_G_path =""
input_B_path =""
input_Depth_path =""
input_NDVI_path =""

R_names = os.listdir(input_R_path)
G_names = os.listdir(input_G_path)
B_names = os.listdir(input_B_path)
Depth_names = os.listdir(input_Depth_path)
NDVI_names = os.listdir(input_NDVI_path)

count = 0
for R_name in R_names:
    image_path = input_R_path + "/" + R_name
    img = np.load(image_path)
    R,_,_,_ = cv2.mean(img)
    R_list.append(R)
        
    
for G_name in G_names:
    image_path = input_G_path + "/" + G_name
    img = np.load(image_path)
    G,_,_,_ = cv2.mean(img)
    G_list.append(G)
    
for B_name in B_names:
    image_path = input_B_path + "/" + B_name
    img = np.load(image_path)
    B,_,_,_ = cv2.mean(img)
    B_list.append(B)
    
for Depth_name in Depth_names:
    image_path = input_Depth_path + "/" + Depth_name
    img = np.load(image_path)
    Depth,_,_,_ = cv2.mean(img)
    Depth_list.append(Depth)


for NDVI_name in NDVI_names:
    image_path = input_NDVI_path + "/" + NDVI_name
    img = cv2.imread(image_path, cv2.IMREAD_UNCHANGED)    
    NDVI,_,_,_ = cv2.mean(img)
    NDVI_list.append(NDVI)
    

#df_final['Region'] = Region_list
df_final['R'] = R_list
df_final['G'] = G_list
df_final['B'] = B_list
df_final['Depth'] = Depth_list
df_final['NDVI'] = NDVI_list

df_final.to_csv('Database Final_Resized Images.csv', header=True, index=False)
```
